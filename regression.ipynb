```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('D:\linear regression sklearn\Linear Regression - Sheet1.csv')
```

```python
df.tail(200)
```

```python
#assinging values to x and y
X = df['X']
Y = df['Y']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
X_train = X_train.to_numpy().reshape(-1, 1)
Y_train = Y_train
```

```python
plt.figure(figsize=(30,30))
plt.scatter(X_train,Y_train)
```

```python
plt.figure(figsize=(30,30))
plt.scatter(X_test , Y_test , color = 'r')
```

```python
model = LinearRegression()

model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test.to_numpy().reshape(-1,1))
print(Y_pred)
```

```python
plt.figure(figsize=(30,30))
plt.scatter(X_test,Y_test)
plt.plot(X_test, Y_pred , color = 'r')
```

```python
train_accuracy = model.score(X_train,Y_train)
train_accuracy
```

```python
test_accuracy = model.score(X_test.to_numpy().reshape(-1,1),Y_test)
test_accuracy
```

```python
test_point = np.array([3.9 , 4.9 , 3.9 ]).reshape(-1,1)
test_point_pred = model.predict(test_point)
print(test_point_pred)
test_point_accuracy = model.score(test_point,test_point_pred)
print(test_point_accuracy)
```

```python
print(model.intercept_)
print(model.coef_)
```
